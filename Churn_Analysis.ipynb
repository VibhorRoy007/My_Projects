```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = pd.read_csv("c.csv")
print("Data loaded. Shape:", x.shape)
```

**Quick Look**

```python

print(x.head())
print(x.info())
print(x.describe())
```

**Clean column names and string data**

```python
x.columns = x.columns.str.strip()
for col in x.select_dtypes(include='object').columns:
    x[col] = x[col].str.strip()
```

**Convert TotalCharges to numeric**

```python
x['TotalCharges'] = pd.to_numeric(x['TotalCharges'], errors='coerce')
x['TotalCharges'].fillna(x['TotalCharges'].median(), inplace=True)
```

**Unique values check**

```python
for col in x.columns:
    print(f"{col}: {x[col].nunique()} unique values")
```

**Categorical columns plots**

```python
cat_cols = x.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=x[col], palette="Set2")
    plt.title(col)
    plt.xticks(rotation=45)
    plt.show()
```

**Numerical columns plots**

```python
num_cols = x.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(x[col], bins=30, kde=True, color='skyblue')
    plt.title(col)
    plt.show()
```

**Boxplots**

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=x[col], color='lightgreen')
    plt.title(col)
    plt.show()
```

**Feature engineering**

```python
x['AvgChargesPerMonth'] = x['TotalCharges'] / x['tenure'].replace(0, np.nan)
x['AvgChargesPerMonth'].fillna(0, inplace=True)
x['SeniorFlag'] = x['SeniorCitizen'].apply(lambda v: 'Yes' if v==1 else 'No')
```

**Churn relationships**

```python
if 'Churn' in x.columns:
    for col in cat_cols:
        if col != 'Churn':
            plt.figure(figsize=(6,4))
            sns.countplot(x=x[col], hue=x['Churn'], palette="Set1")
            plt.title(f"{col} vs Churn")
            plt.xticks(rotation=45)
            plt.show()
```

```python
for col in num_cols:
        if col != 'customerID':
            plt.figure(figsize=(6,4))
            sns.boxplot(x=x['Churn'], y=x[col], palette="Set3")
            plt.title(f"{col} vs Churn")
            plt.show()
```

**Correlation heatmap**

```python
plt.figure(figsize=(10,6))
sns.heatmap(x[num_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Pairplot**

```python
selected_cols = [col for col in num_cols if col in ['tenure', 'MonthlyCharges', 'TotalCharges']]
if len(selected_cols) > 1:
    sns.pairplot(x[selected_cols + ['Churn']], hue='Churn', diag_kind='kde')
    plt.show()
```

**Tenure grouping**

```python
bins = [0,12,24,36,48,60,72]
labels = ['0-12','12-24','24-36','36-48','48-60','60-72']
x['TenureGroup'] = pd.cut(x['tenure'], bins=bins, labels=labels, right=False)
plt.figure(figsize=(6,4))
sns.countplot(x=x['TenureGroup'], hue=x['Churn'], palette="pastel")
plt.title("Churn by Tenure Group")
plt.show()
```

**Payment method analysis**

```python
plt.figure(figsize=(8,5))
sns.countplot(x=x['PaymentMethod'], hue=x['Churn'], palette="muted")
plt.xticks(rotation=45)
plt.title("Churn by Payment Method")
plt.show()
```

**Contract type analysis**

```python
plt.figure(figsize=(6,4))
sns.countplot(x=x['Contract'], hue=x['Churn'], palette="bright")
plt.title("Churn by Contract Type")
plt.show()
```

**Internet service analysis**

```python
plt.figure(figsize=(6,4))
sns.countplot(x=x['InternetService'], hue=x['Churn'], palette="Set2")
plt.title("Churn by Internet Service")
plt.show()
```

**Online services analysis**

```python
online_services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for col in online_services:
    if col in x.columns:
        plt.figure(figsize=(6,4))
        sns.countplot(x=x[col], hue=x['Churn'], palette="cool")
        plt.title(f"{col} vs Churn")
        plt.xticks(rotation=45)
        plt.show()
```

**Gender, Partner, Dependents analysis**

```python
for col in ['gender', 'Partner', 'Dependents']:
    if col in x.columns:
        plt.figure(figsize=(6,4))
        sns.countplot(x=x[col], hue=x['Churn'], palette="Set1")
        plt.title(f"{col} vs Churn")
        plt.show()
```

**Monthly charges vs churn**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=x['Churn'], y=x['MonthlyCharges'], palette="Set2")
plt.title("Monthly Charges vs Churn")
plt.show()
```

**Total charges vs churn**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=x['Churn'], y=x['TotalCharges'], palette="Set3")
plt.title("Total Charges vs Churn")
plt.show()
```

**Average charges per month vs churn**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=x['Churn'], y=x['AvgChargesPerMonth'], palette="Set2")
plt.title("Avg Charges per Month vs Churn")
plt.show()
```

**Senior flag vs churn**

```python
plt.figure(figsize=(6,4))
sns.countplot(x=x['SeniorFlag'], hue=x['Churn'], palette="Set1")
plt.title("Senior Citizen vs Churn")
plt.show()
```

**Group tenure vs monthly charges**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=x['TenureGroup'], y=x['MonthlyCharges'], palette="pastel")
plt.title("Monthly Charges by Tenure Group")
plt.show()
```

**Group tenure vs total charges**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=x['TenureGroup'], y=x['TotalCharges'], palette="pastel")
plt.title("Total Charges by Tenure Group")
plt.show()
```
